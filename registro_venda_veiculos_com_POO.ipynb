Classe pai veiculos

```python
class Veiculos:
    def __init__(self, marca, modelo, ano, cor, preco) -> None:
        self._marca = marca
        self._modelo = modelo
        self._ano = ano
        self._cor = cor
        self._preco = preco
    
    
    @property
    def marca(self):
        return self._marca
    
    @marca.setter
    def marca(self, valor):
        self._marca = valor
        
    @property
    def modelo(self):
        return self._modelo
    
    @modelo.setter
    def modelo(self, valor):
        self._modelo = valor
    
    @property
    def ano(self):
        return self._ano
    
    @ano.setter
    def ano(self, valor):
        self._ano = valor

    @property
    def cor(self):
        return self._cor

    @cor.setter
    def cor(self, valor):
        self._cor = valor

    @property
    def preco(self):
        return self._preco

    @preco.setter
    def preco(self, valor):
        self._preco = valor

        
    def informacoes(self):
        return f"Marca: {self._marca}, Modelo: {self._modelo}, Ano: {self._ano}, Cor: {self._cor}, Preço: R${self._preco}"
        
```

Classe filha carro

```python
class Carro(Veiculos):
    def __init__(self, marca, modelo, ano, cor, preco, num_portas) -> None:
        super().__init__(marca, modelo, ano, cor, preco)
        self._num_portas = num_portas
    
    
    @property
    def num_portas(self):
        return self._num_portas
    
    @num_portas.setter
    def num_portas(self, valor):
        self._num_portas = valor
        

    def informacoes(self):
        info = super().informacoes()
        return f"Carro, {info}, Número de portas: {self._num_portas}"
```

Classe filha moto

```python
class Moto(Veiculos):
    def __init__(self, marca, modelo, ano, cor, preco, cilindradas) -> None:
        super().__init__(marca, modelo, ano, cor, preco)
        self._cilindradas = cilindradas
        
        
    @property
    def cilindradas(self):
        return self._cilindradas
    
    @cilindradas.setter
    def cilindradas(self, valor):
        self._cilindradas = valor
        
        
    def informacoes(self):
        info = super().informacoes()
        return f"Moto, {info}, Cilindradas: {self._cilindradas}"
```

Classe filha caminhonete

```python
class Caminhonete(Veiculos):
    def __init__(self, marca, modelo, ano, cor, preco, volume_cacamba) -> None:
        super().__init__(marca, modelo, ano, cor, preco)
        self._volume_cacamba = volume_cacamba
        
        
    @property
    def volume_cacamba(self):
        return self._volume_cacamba
    
    @volume_cacamba.setter
    def volume_cacamba(self, valor):
        self._volume_cacamba = valor
        
        
    def informacoes(self):
        info = super().informacoes()
        return f"Caminhonete, {info}, Volume da caçamba: {self._volume_cacamba}"
    
```

Classe do comprador

```python
class Comprador:
    def __init__(self, nome, cpf, data_compra):
        self.__nome = nome
        self.__cpf = cpf
        self.__data_compra = data_compra
        
        
    @property
    def nome(self):
        return self.__nome
    
    @property
    def cpf(self):
        return self.__cpf
    
    @property
    def data_compra(self):
        return self.__data_compra
    
    
    def informacoes(self):
        return f"Cliente: {self.__nome}, CPF: {self.__cpf}, Data da compra: {self.__data_compra}"
```

Classe da concessionária

```python
class Concessionaria:
    def __init__(self, nome, vendedor):
        self.__nome = nome
        self.__vendedor = vendedor
        
    
    @property
    def nome(self):
        return self.__nome
    
    @property
    def vendedor(self):
        return self.__vendedor
    
    
    def informacoes(self):
        return f"Concessionária: {self.__nome}, Vendedor: {self.__vendedor}"
```

Funções para validar entradas sobre o cliente

```python
# Verifica se o CPF tem exatamente 11 dígitos e contém apenas números
def validar_cpf(cpf):
    if cpf.isdigit() and len(cpf) == 11:
        return True
    
    else:
        return False
    

# Formata o CPF para o formato xxx.xxx.xxx-xx
def formatar_cpf(cpf):
    return f"{cpf[:3]}.{cpf[3:6]}.{cpf[6:9]}-{cpf[9:]}"    
    

# Valida a data no formato dd/mm/yyyy
def validar_data(data):
    if len(data) != 10:
        return False
    
    # Verifica se o terceiro e sexto caracteres são barras
    if data[2] != '/' or data[5] != '/':
        return False
    
    # Verifica se dia, mês e ano são numéricos
    dia = data[:2]
    mes = data[3:5]
    ano = data[6:]
    
    if not (dia.isdigit() and mes.isdigit() and ano.isdigit()):
        return False
    
    # Verifica se dia, mês e ano estão são válidos dentro do calendário
    dia = int(dia)
    mes = int(mes)
    ano = int(ano)
    
    if mes < 1 or mes > 12:
        return False
    
    if dia < 1 or dia > 31:
        return False
    
    return True
```

Fuções para registrar e ler em arquivo txt

```python
# Registra os dados em arquivo txt
def salvar_venda(concessionaria, comprador, veiculo):
    arquivo_nome = concessionaria.nome.replace(" ", "_") + ".txt"
    with open(arquivo_nome, "a") as arquivo:
        venda_info = concessionaria.informacoes() + ' | ' + comprador.informacoes() + ' | ' + veiculo.informacoes() + "\n"
        arquivo.write(venda_info)
        arquivo.write("="*50 + "\n")
    print(f"Venda registrada no arquivo {arquivo_nome}")

# Lê os dados em arquivo txt
def exibir_vendas(concessionaria):
    arquivo_nome = concessionaria.nome.replace(" ", "_") + ".txt"
    try:
        with open(arquivo_nome, "r") as arquivo:
            print(arquivo.read())
    except FileNotFoundError:
        print(f"Nenhum registro encontrado para a concessionária {concessionaria.nome}")
```

Menus para registro/busca de vendas

```python
menu1 = """Digite a operação desejada:

            [1] Registrar nova venda
            [2] Buscar venda realizada
            [3] Sair
            
            """

menu2 = """Selecione a concessionária:

            [1] Recreio Veículos
            [2] Barra Veículos
            [3] Tijuca Veículos
            
            """

menu3 = """Selecione o tipo de veículo:

            [1] Carro
            [2] Moto
            [3] Caminhonete
            
            """    
    
menu4 = """Selecione a marca do veículo:

            [1] Toyota
            [2] Ford
            [3] Chevrolet
            [4] Volkswagen  
             
            """
            
menu5 = """Selecione a marca do veículo:
            
            [1] Honda
            [2] Yamaha
            [3] Suzuki
            [4] Kawasaki
            
            """         
            
menu6 = """Digite a operação desejada:
            
            [1] Salvar registro
            [2] Editar registro
            [3] Descartar registro
            
            """           
             
menu7 = """Qual campo deseja editar:
            
            [1] Tipo de veículo
            [2] Marca
            [3] Modelo
            [4] Ano
            [5] Cor
            [6] Preço
            [7] Número de portas (Se tipo de veículo for carro)
            [8] Cilindradas (Se tipo de veículo for moto)
            [9] Capacidade da caçamba (Se tipo de veículo for caminhonete)
            [0] Voltar ao menu anterior
            
            ** Por questões de segurança dados da concerssionária, vendedor e comprador não podem ser editados,
            se for necessário descarte as alterações e crie um novo registro **
            
            """              
```

Registro/busca das vendas 

```python
while True:
    
    opcao = input(menu1)
    
    if opcao == '1': # Registrar nova venda
        
        while True:
            concessionaria_opcao = input(menu2)
            vendedor_opcao = input("Digite o nome do vendedor: ")
            vendedor_opcao = vendedor_opcao.title()
        
            if concessionaria_opcao == '1': # Recreio Veículos
                concessionaria = Concessionaria('Recreio Veículos', vendedor_opcao)
                break
            
            elif concessionaria_opcao == '2': # Barra Veículos
                concessionaria = Concessionaria('Barra Veículos', vendedor_opcao)
                break
        
            elif concessionaria_opcao == '3': # Tijuca Veículos
                concessionaria = Concessionaria('Tijuca Veículos', vendedor_opcao)
                break
        
            else:
                print("Opção inválida.")
    
    
        nome_comprador = input("Digite o nome do comprador: ")
        nome_comprador = nome_comprador.title()

        while True:
            cpf_comprador = input("Digite o CPF do comprador (apenas números): ")
            if validar_cpf(cpf_comprador):
                cpf_comprador_formatado = formatar_cpf(cpf_comprador)
                break
            
            else:
                print("Formato do CPF inválido.")
                
        while True:
            data_compra = input("Digite a data da compra (no formato dd/mm/aaaa): ")
            if validar_data(data_compra):
                break
            else:
                print("Formato da data inválido.")        
        
        # Cria variável para todas as descrições de comprador usando a classe
        comprador = Comprador(nome_comprador, cpf_comprador_formatado, data_compra)
        
                
        while True:
            tipo_veiculo = input(menu3)
            
            if tipo_veiculo == '1': # Carro
                marca_veiculo = input(menu4)
                modelo_veiculo = input("Digite o modelo do veículo: ")
                ano_veiculo = input("Digite o ano do veículo: ")
                cor_veiculo = input("Digite a cor do veículo: ")
                preco_veiculo = input("Digite o preço do veículo: ")
                num_portas_carro = input("Digite o número de portas do veículo: ")
                
                if marca_veiculo == '1': # Toyota
                    marca_veiculo = "Toyota"
                    
                elif marca_veiculo == '2': # Ford
                    marca_veiculo = "Ford"
                    
                elif marca_veiculo == '3': # Chevrolet
                    marca_veiculo = "Chevrolet"
                    
                elif marca_veiculo == '4': # Volkswagen 
                    marca_veiculo = "Volkswagen"
                    
                else:
                    print("Opção inválida.")    
                    continue            
                
                # Cria variável com todas as descrições de tipo_veiculo para carro usando a classe 
                carro = Carro(marca_veiculo, modelo_veiculo, ano_veiculo, cor_veiculo, preco_veiculo, num_portas_carro)
                veiculo = carro
                
            
            elif tipo_veiculo == '2': # Moto
                marca_veiculo = input(menu5)
                modelo_veiculo = input("Digite o modelo do veículo: ")
                ano_veiculo = input("Digite o ano do veículo: ")
                cor_veiculo = input("Digite a cor do veículo: ")
                preco_veiculo = input("Digite o preço do veículo: ")
                cilindradas_moto = input("Digite as cilindradas do veículo: ")

                if marca_veiculo == '1': # Honda
                    marca_veiculo = "Honda"
                    
                elif marca_veiculo == '2': # Yamaha
                    marca_veiculo = "Yamaha"
                    
                elif marca_veiculo == '3': # Suzuki
                    marca_veiculo = "Suzuki"
                    
                elif marca_veiculo == '4': # Kawasaki 
                    marca_veiculo = "Kawasaki"
                    
                else:
                    print("Opção inválida.")
                    continue
                
                # Cria variável com todas as descrições de tipo_veiculo para moto usando a classe
                moto = Moto(marca_veiculo, modelo_veiculo, ano_veiculo, cor_veiculo, preco_veiculo, cilindradas_moto)
                veiculo = moto
                
                
            elif tipo_veiculo == '3': # Caminhonete
                marca_veiculo = input(menu4)
                modelo_veiculo = input("Digite o modelo do veículo: ")
                ano_veiculo = input("Digite o ano do veículo: ")
                cor_veiculo = input("Digite a cor do veículo: ")
                preco_veiculo = input("Digite o preço do veículo: ")
                cacamba_caminhonete = input("Digite a capacidade da caçamba do veículo em litros: ")
                
                if marca_veiculo == '1': # Toyota
                    marca_veiculo = "Toyota"
                    
                elif marca_veiculo == '2': # Ford
                    marca_veiculo = "Ford"
                    
                elif marca_veiculo == '3': # Chevrolet
                    marca_veiculo = "Chevrolet"
                    
                elif marca_veiculo == '4': # Volkswagen 
                    marca_veiculo = "Volkswagen"
                  
                    
                else:
                    print("Opção inválida.") 
                    continue 
                
                # Cria variável com todas as descrições de tipo_veiculo para caminhonete usando a classe
                caminhonete = Caminhonete(marca_veiculo, modelo_veiculo, ano_veiculo, cor_veiculo, preco_veiculo, cacamba_caminhonete)
                veiculo = caminhonete
                
                                                       
            else: 
                print("Opção inválida.")  
                continue
            break        
        
        
        # Loop para confirmação de gravação, edição ou descarte dos dados
        while True:
            
            # Exibir informações antes de gravar
            print("\nDados da Venda:")
            print(concessionaria.informacoes())
            print(comprador.informacoes())
            print(veiculo.informacoes())
            
            add_edit = input(menu6)
            
            if add_edit == '2': # Editar registro
                editar = input(menu7)

                if editar == '1': # Tipo de veículo
                    novo_tipo = input(menu3)
                    
                    if novo_tipo == '1': # Novo tipo é Carro
                        num_portas = input("Digite o número de portas do veículo: ")
                        veiculo = Carro(marca_veiculo, modelo_veiculo, ano_veiculo, cor_veiculo, preco_veiculo, num_portas)
    
                    elif novo_tipo == '2': # Novo tipo é Moto
                        cilindradas = input("Digite as cilindradas do veículo: ")
                        veiculo = Moto(marca_veiculo, modelo_veiculo, ano_veiculo, cor_veiculo, preco_veiculo, cilindradas)
    
                    elif novo_tipo == '3': # Novo tipo é Caminhonete
                        volume_cacamba = input("Digite a capacidade da caçamba do veículo em litros: ")
                        veiculo = Caminhonete(marca_veiculo, modelo_veiculo, ano_veiculo, cor_veiculo, preco_veiculo, volume_cacamba)
    
                    else:
                        print("Opção inválida.")
                    
                    
                elif editar == '2': # Marca
                    if carro or caminhonete:
                        nova_marca = input(menu4)
                        veiculo.marca = nova_marca
                        
                    elif moto:
                        nova_marca = input(menu5)
                        veiculo.marca = nova_marca
                    
                elif editar == '3': # Modelo
                    novo_modelo = input("Digite o novo modelo do veículo: ")
                    veiculo.modelo = novo_modelo
                    
                elif editar == '4': # Ano
                    novo_ano = input("Digite o novo ano do veículo: ")
                    veiculo.ano = novo_ano
                    
                elif editar == '5': # Cor
                    nova_cor = input("Digite a nova cor do veículo: ")
                    veiculo.cor = nova_cor
                    
                elif editar == '6': # Preço
                    novo_preco = input("Digite o novo preço do veículo: ")
                    veiculo.preco = novo_preco
                    
                elif editar == '7': # Número de portas (Se tipo de veículo for carro)
                    if isinstance(veiculo, Carro): # Verifica se objeto é uma instância da classe Carro
                        novas_portas = input("Digite o novo número de portas: ")
                        veiculo.num_portas = novas_portas
                        
                    else:
                        print("Operação não permitida para esse tipo de veículo.")
                    
                elif editar == '8': # Cilindradas (Se tipo de veículo for moto)
                    if isinstance(veiculo, Moto): # Verifica se objeto é uma instância da classe Moto
                        novas_cilindradas = input("Digite as novas cilindradas: ")
                        veiculo.cilindradas = novas_cilindradas
                    
                    else:
                        print("Operação não permitida para esse tipo de veículo.")
                    
                elif editar == '9': # Capacidade da caçamba (Se tipo de veículo for caminhonete)
                    if isinstance(veiculo, Caminhonete): # Verifica se objeto é uma instância da classe Caminhonete
                        nova_cacamba = input("Digite a nova capacidade da caçamba: ")
                        veiculo.cacamba = nova_cacamba
                        
                    else:
                        print("Operação não permitida para esse tipo de veículo.")
                    
                elif editar == '0': # Voltar ao menu anterior
                    break
                    
                else:
                    print("Opção inválida.")
             
                
            elif add_edit == '3': # Descartar registro
                del concessionaria
                del comprador
                del veiculo
                print("Registro descartado.")
                break
             
                
            elif add_edit == '1': # Salvar registro
                salvar_venda(concessionaria, comprador, veiculo)
                break

            else:
                print("Opção inválida.")
                
            
    elif opcao == '2': # Buscar venda realizada

        buscar_venda = input(menu2)
        if buscar_venda == '1': # Recreio Veículos
            print("Registro de vendas de Recreio Veículos: \n")
            concessionaria = Concessionaria('Recreio Veículos', '')
            exibir_vendas(concessionaria)
            break 
        
        elif buscar_venda == '2': # Barra Veículos
            print("Registro de vendas de Barra Veículos: \n")
            concessionaria = Concessionaria('Barra Veículos', '')
            exibir_vendas(concessionaria)
            break
        
        elif buscar_venda == '3': # Tijuca Veículos
            print("Registro de vendas de Tijuca Veículos: \n")
            concessionaria = Concessionaria('Tijuca Veículos', '')
            exibir_vendas(concessionaria)
            break
        
        else:
            print("Opção inválida.")
            continue

            
    elif opcao == '3': # Sair
        print("Finalizando Terminal")
        break
    
    
    else:
        print("Opção inválida, por favor insira uma das opções informadas.")
```
